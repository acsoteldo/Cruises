```python
import pandas as pd
data = pd.read_csv('cruise_dataset.csv')
```

```python
data.head()
data.info()
data.describe()
```

```python
data.dropna()
data.drop_duplicates()
```

```python
data['Ship_name'].fillna('unknown', inplace=True)
data['Cruise_line'].fillna('unknown', inplace=True)
```

```python
data.fillna('0', inplace=True)  # Replace NaN with '0'
```

```python
data.rename(columns={'passengers': 'Passengers'}, inplace=True)
data.rename(columns={'length': 'Length'}, inplace=True)
data.rename(columns={'cabins': 'Cabins'}, inplace=True)
data.rename(columns={'passenger_density': 'Passenger_density'}, inplace=True)
data.rename(columns={'crew': 'Crew'}, inplace=True)
```

```python
data.to_csv('cruise_dataset CLEANED.csv', index=False)
```
